Exploratory Data Analysis 


Gender

```python
df.columns
```

```python
ax = sns.countplot(x= 'Gender' , data = df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['Gender'], as_index = False)
['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen= df.groupby(['Gender'], as_index = False)
['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

From above graghs we can see that most of the buyers are females and even the purchasing power of females are greater than men.


Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender' )
for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x = 'Age Group', y= 'Amount', data = sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 years female.

State

```python
#Total number of orders from 10 states
df.groupby(['State'], as_index= False)['Orders']
.sum().sort_values(by = 'Orders', ascending = True)
```

```python
orders_state =df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
```

```python
sns.barplot( x= 'State', y = 'Orders', data = orders_state)
sns.set(rc = {'figure.figsize':(30,8)})
```

```python
#Total amount/sales from top 10 states


sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot( x= 'State', y = 'Amount', data= sales_state)
```

From the above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra,Karnataka respectively.

Marital Status

```python
ax = sns.countplot(data = df, x= 'Marital_Status')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.set(rc= {'figure.figsize': (6,5)})
```

```python
sns.barplot( data = sales_state, x= 'Marital_Status', y= 'Amount', hue= 'Gender' )
```

From the above graphs we can see that most of the buyers are married(women) and they have high purchasing power.


Occupation

```python
ax = sns.countplot(data = df, x= 'Occupation')
for i in ax.containers:
    ax.bar_label(i)
sns.set(rc = {'figure.figsize':(30,5)})
```

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot( x= 'Occupation', y = 'Amount', data = sales_occupation )
```

From above grapghs we can see that most of the buyers are working in IT sector, Healthcare and Aviation sector . 

Product Category

```python
ax = sns.countplot(data = df, x= 'Product_Category')
for i in ax.containers:
    ax.bar_label(i)
sns.set(rc = {'figure.figsize':(30,5)})
```

```python
sales_productwise = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
```

```python
sns.set(rc = {'figure.figsize' : (35,5)})
sns.barplot( x= 'Product_Category', y = 'Amount', data = sales_productwise )
```

From the above graphs we can see that most of the sold products are from Food,clothing and apparel, electronics and gadgets and footwear and shoes.

```python
#To find top 10 selling products
sales_topproducts = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
```

```python
sns.barplot( x= 'Product_ID', y = 'Orders', data = sales_topproducts )
```

Conclusion:
Married women age group 26-35 from Uttar Pradesh, Maharashtra and Karnataka working in the IT sector, Healthcare and Aviation are more likely to buys products from Food , Clothing and Electronics category.
